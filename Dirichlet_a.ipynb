**Solución a la ecuación de Dirichlet.**

La solución de la ecuación:
$$\frac{∂^2 u}{∂x^2} + \frac{∂^2u}{∂y^2} = 0, \hspace{1.5cm} 0<x<1, \hspace{1.5cm} 0<y<1$$
Con condiciones de frontera: 
$$a) \hspace{0.5cm}u(0,y) = u(1,y) = 0; u(x,0)=100; u(x,1)=200$$
Está dada en términos de la serie de Fourier por la expresión:
$$u(x,y) = \sum_{n=1}^\infty (A_n cosh (n\pi y)+ B_n sinh(n\pi y))sin(n\pi x),$$
donde:
$$A_n = \frac{200[1-(-1)^n]}{n\pi}$$
$$B_n = \frac{200[1-(-1)^n][2-cosh(n\pi)]}{n\pi sinh(n\pi)}$$



```python
from pylab import *
x = linspace(0,1.0,1000)
y = linspace(0,1.0,1000)

def u_x_y(x,y):
    N = 10
    f = zeros((len(x),len(y)))
    for i in range(len(y)):
        for n in range(1,N):
          A = (200*(1-(-1)**(n)))/(n*pi)
        
          B = (200*(1-(-1)**(n))*(2-cosh(n*pi)))/((n*pi)*sinh(n*pi))
          f[:,i] += sin(n*pi*x)*(A*cosh(n*pi*y[i])+B*sinh(n*pi*y[i]))
    return f


 
#%matplotlib notebook
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(12,10))
ax = fig.add_subplot(111, projection="3d")

X,Y = np.meshgrid(x,y)

U = u_x_y(x,y)

plt.xlabel('$y$'); plt.ylabel('$x$'); 
plt.title("u(x,y)")
ax.plot_surface(X, Y, U, cmap="copper");
```
